```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import precision_recall_curve, confusion_matrix

# Data from the table
security_mechanisms = ["FIM", "Suricata", "Command Detection", "VirusTotal"]
initial_precision = [87.2, 82.5, 85.6, 91.3]
final_precision = [93.5, 90.1, 92.4, 96.8]
initial_recall = [80.4, 75.3, 78.2, 88.5]
final_recall = [89.2, 86.7, 88.9, 94.2]

# Creating synthetic confusion matrices before and after fine-tuning (simplified for visualization)
conf_matrix_before = np.array([[85, 15], [20, 80]])  # Before fine-tuning
conf_matrix_after = np.array([[92, 8], [10, 90]])  # After fine-tuning

# Figure X: Confusion Matrix Before & After Fine-Tuning
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Before Fine-Tuning
sns.heatmap(conf_matrix_before, annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Confusion Matrix - Before Fine-Tuning")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")

# After Fine-Tuning
sns.heatmap(conf_matrix_after, annot=True, fmt="d", cmap="Blues", ax=axes[1])
axes[1].set_title("Confusion Matrix - After Fine-Tuning")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

plt.tight_layout()
plt.show()
```

```python
# Generate Precision-Recall Curve for Initial and Final Performance

# Simulated precision-recall data
recall_range = np.linspace(0, 1, 100)
initial_precision_curve = np.interp(recall_range, [0, 1], [min(initial_precision)/100, max(initial_precision)/100])
final_precision_curve = np.interp(recall_range, [0, 1], [min(final_precision)/100, max(final_precision)/100])

plt.figure(figsize=(8, 6))
plt.plot(recall_range, initial_precision_curve, linestyle='--', label="Initial Precision-Recall", color='red')
plt.plot(recall_range, final_precision_curve, linestyle='-', label="Final Precision-Recall", color='blue')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Figure Y: Precision-Recall Curve Before & After Fine-Tuning")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate Detection Accuracy Graph

x_labels = ["FIM", "Suricata", "Command Detection", "VirusTotal"]
x = np.arange(len(x_labels))

plt.figure(figsize=(8, 6))
plt.bar(x - 0.2, initial_precision, width=0.4, label="Initial Precision", color='red', alpha=0.6)
plt.bar(x + 0.2, final_precision, width=0.4, label="Final Precision", color='blue', alpha=0.6)

plt.xlabel("Security Mechanism")
plt.ylabel("Precision (%)")
plt.title("Figure Z: Detection Accuracy Improvement After Fine-Tuning")
plt.xticks(ticks=x, labels=x_labels)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
